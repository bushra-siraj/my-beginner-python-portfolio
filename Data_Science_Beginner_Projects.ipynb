2. To do List Manager
Concepts: Lists, append, remove, slicing
Challenge: Let users add, remove, and view items in their to do list.


```python
todo_list = []

while True:
  print("Please select one of the following options")
  print("1. Add a new item")
  print("2. Remove an item")
  print("3. View items")
  print("4. Quit")
  choice = input("Enter your choice: ")

  if choice == "1":
    new_item = input("Please enter a to-do item: ")
    todo_list.append(new_item)
    print(f"Item '{new_item}' added to the to-do list.")

  elif choice == "2":
    if not todo_list:
      print("Your to-do list is empty. Nothing to remove.")

    else:
     print("To-Do List:")
    for item in todo_list:
      print(f"{item}")

      item_number_to_remove = int(input("Enter the item number to remove: "))
      item_to_remove = item_number_to_remove - 1

      if 0 <= item_to_remove < len(todo_list):
          removed_item = todo_list.pop(item_to_remove)
          print(f"Item '{removed_item}' removed from the to-do list.")
      else:
          print(f"Invalid item number. Item {item_number_to_remove} not found.")

  elif choice == "3":
    print("Your To-Do List:")
    if not todo_list:
      print("Your to-do list is empty!")
    else:
      for item in todo_list:
        print(item)
  elif choice == "4":
    print("Goodbye!")
    break

  else:
    print("Invalid input. Please try again.")
```

3. City Search
Concepts: for loops, list membership
Challenge: Check if a city is in a predefined list of clean cities.

```python
clean_cities = ["Islamabad", "Karachi", "Lahore", "Peshawar", "Muree"]

city_to_search = input("Enter a city name to check if it's in the list of clean cities: ")

found = False
for city in clean_cities:
    if city == city_to_search:
        print(f"{city_to_search} is in the list of clean cities!")
        found = True
        break

if not found:
    print(f"{city_to_search} is not in the list of clean cities.")
```

4. Expense Logger
Concepts: Lists/dictionary, input, math
Challenge: Let users enter daily expenses and calculate the total.

```python
expenses = []

def add_expense():
  while True:
    expense_input = input("Enter the expense amount:")
    break

  expense_amount = float(expense_input)
  expenses.append(expense_amount)
  print(f"Expense of {expense_amount:} added.")

def view_expenses():
  if not expenses:
    print("No expenses seen yet.")
  else:
    print("View Expenses:")
    for expense in expenses:
     expense = float(expense)
     print(f"{expense:}")

def calculate_total():
  if not expenses:
    print("No expenses to calculate.")
  else:
    total_expense = sum(expenses)
    print(f"Total Expenses: {total_expense:}")

print("Welcome to the Expense Logger!")
while True:
  print("Choose an action:")
  print("1. Add expense")
  print("2. View expenses")
  print("3. Calculate total")
  print("4. Quit")

  choice = input("Enter your choice (1-4): ")

  if choice == '1':
    add_expense()
  elif choice == '2':
    view_expenses()
  elif choice == '3':
    calculate_total()
  elif choice == '4':
    print("Exiting Expense Logger. Goodbye!")
    break
  else:
    print("Invalid choice. Please enter a number between 1 and 4.")
```

5. BMI Calculator
Concepts: Math, input, if/else
Challenge: Input weight and height, calculate BMI, and classify it.

```python
height = float(input("Enter your height in meters: "))
weight = float(input("Enter your weight in kilograms: "))

bmi = weight / (height ** 2)

if bmi < 18.5:
    print("Underweight")
elif 18.5 <= bmi < 24.9:
    print("Normal weight")
else:
    print("Overweight")
```

6. Quiz App
Concepts: Functions, dictionaries, conditionals
Challenge: Create a quiz with 3–5 questions stored in a dictionary. Ask the user each question and keep score.

```python
questions = [
    {
        'question': 'How old is Python?',
        'options': {
            'a': '20 years',
            'b': '30 years',
            'c': '40 years',
        },
        'answer': 'b'
    },
    {
        'question': 'What does CPU stand for?',
        'options': {
            'a': 'Central Processing Unit',
            'b': 'Computer Personal Unit',
            'c': 'Central Personal Unit',
        },
        'answer': 'a'
    },
    {
        'question': 'Who is the president of Pakistan',
        'options': {
            'a': 'Shabaz Shareef',
            'b': 'Nawaz Shareef',
            'c': 'Imran Khan',
        },
        'answer': 'a'
    }
]

score = 0
print("Welcome to the Quiz App!")

for q in questions:
    print(q['question'])

    for option_key, option_value in q['options'].items():
        print(option_key, option_value)

    user_answer = input("Enter your answer (a, b, or c): ")

    if user_answer == q['answer']:
        print("Correct!")
        score += 1
    else:
        print("Incorrect.")

print("Quiz finished.")
print(f'Your final score is: {score}')
```

8. Bank Account Simulator
Concepts: Class, methods, state
Challenge: Create a BankAccount class with methods to deposit, withdraw, and check_balance.

```python
class BankAccount:
  def __init__(self):
    self.balance = 0
    print("A new bank account has been created.")

  def deposit(self, amount):
    if amount > 0:
      self.balance += amount
      print(f"Deposit successful. Current balance: ${self.balance:}")
    else:
      print("Invalid deposit amount.")

  def withdraw(self, amount):
    if amount > 0 and amount <= self.balance:
      self.balance -= amount
      print(f"Withdrawal successful. Remaining balance: ${self.balance:}")
    elif amount > self.balance:
      print("Insufficient funds.")
    else:
      print("Invalid withdrawal amount.")

  def check_balance(self):
    print(f"Current balance: ${self.balance:}")
```

```python
my_account = BankAccount()
my_account.deposit(10000000000000)
my_account.withdraw(50006789)
my_account.check_balance()
```

9. console-based Student Portal
Concepts: class, methods
Challenge: that allows users to:
Add new students
View student details
Update grades
Calculate average grade
Exit the program

```python
class Student:
    def __init__(self, name, student_id, grade):
        self.name = name
        self.student_id = student_id
        self.grade = grade

    def students_details(self):
        print(f"Student id: {self.student_id}")
        print(f"Name: {self.name}")
        print(f"Grade: {self.grade}")

    def update_grade(self, new_grade):
        self.grade = new_grade
        print(f"Grade for {self.name} (id: {self.student_id}) updated to {self.grade}.")

students = []

def add_student():
    name = input("Enter student name: ")
    student_id = input("Enter student ID: ")

    new_student = Student(name, student_id)
    students.append(new_student)
    print(f"Student '{name}' (ID: {student_id}) added successfully.")
    print(new_student)

def view_students():
    if not students:
        print("No students in the portal yet.")
    else:
        print("Student Details")
        for student in students:
            student.students_details()

def update_student_grade():
    student_id = input("Enter the ID of the student to update grades: ")
    found_student = None

    for student in students:
        if student.student_id == student_id:
            found_student = student
            break

    if found_student:
        new_grade = float(input("Enter the new grade for the student: "))
        found_student.update_grade(new_grade)
    else:
        print(f"Student with ID {student_id} not found.")

def calculate_average_grade():
    grades = []
    for student in students:
        if student.grade is not None:
            grades.append(student.grade)

    if not grades:
        print("No grades available to calculate average.")
    else:
        average = sum(grades) / len(grades)
        print(f"Average grade of all students: {average:}")

print("Welcome to the Student Portal!")
while True:
      print("Choose an action:")
      print("1. Add new student")
      print("2. View students")
      print("3. Update student grade")
      print("4. Calculate average grade")
      print("5. Exit")

      choice = input("Enter your choice (1-5): ")

      if choice == '1':
        add_student()
      elif choice == '2':
        view_students()
      elif choice == '3':
        update_student_grade()
      elif choice == '4':
        calculate_average_grade()
      elif choice == '5':
        print("Exiting Student Portal. Goodbye!")
        break
      else:
        print("Invalid choice. Please enter a number between 1 and 5.")
```

10. Snack Vending Machine
Concepts: function, while loops, classes
Challenge: Simulate a vending machine that sells snacks. The user can:
View available snacks
Select a snack
Check balance
Exit the machine

```python
class Snack:
    def __init__(self, name, price):
        self.name = name
        self.price = price

available_snacks = [
    Snack("Chips", 10.0),
    Snack("Candy", 100.0),
    Snack("juice", 200.0),
    Snack("Cookies", 20.0)
]

balance = [500.0]

def display_snacks():
    print("Available Snacks")
    snack_number = 1
    for snack in available_snacks:
        print(f"{snack_number}, {snack.name} - ${snack.price:}")
        snack_number += 1

def purchase_snack(snack_index, balance_list):
    if 0 <= snack_index < len(available_snacks):
        selected_snack = available_snacks[snack_index]
        snack_price = selected_snack.price

        if balance_list[0] >= snack_price:
            balance_list[0] -= snack_price
            print(f"You purchased {selected_snack.name}:")
        else:
            print(f"Insufficient funds.")
    else:
        print("Invalid snack selection.")

def check_balance(balance_list):
    print(f"Your current balance is: ${balance_list[0]:}")

print("Welcome to the Snack Vending Machine!")

while True:
    print("Choose an action:")
    print("1. View snacks")
    print("2. Select snack to purchase")
    print("3. Check balance")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        display_snacks()
    elif choice == '2':
        display_snacks()
        while True:
                snack_number = int(input("Enter the number of the snack you want to purchase: "))
                purchase_snack(snack_number - 1, balance)
                break
                print("Invalid input. Please enter a number.")

    elif choice == '3':
        check_balance(balance)
    elif choice == '4':
        print("Exiting Vending Machine. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 4.")
```
